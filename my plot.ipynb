```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('final_data_lag_days_1.csv')
```

```python
#data['flag'] = np.where(data['pos_pol']>=data['neg_pol'],'positive','negative')
data['flag'] = np.where(data['compound']>=0.4,'positive','negative')
```

```python
count = data['flag'].value_counts()/len(data)*100 + 0.5
count = count.astype(int)
```

```python
# 生成测试数据
x = count.index
y = count.values

# 绘制柱状图
plt.bar(x, y,color=['r','g','b', 'c', 'm', 'y'])
# 循环，为每个柱形添加文本标注

# 居中对齐
for xx, yy in zip(x,y):
    plt.text(xx, yy+0.1, str(yy)+'%', ha='center')

plt.ylabel('Count', fontsize=12)
plt.xlabel('Sentiment', fontsize=12)
plt.show();

# 显示图形
plt.show();
```
